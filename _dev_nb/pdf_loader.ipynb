```python
async with PdfLoader() as loader:
    output_path = Path("_dev_nb/output_data/pdf_loader")
    for k, data in sample_pdfs.items():
        docs = await loader.load_document(file_path=data["pdf"])
        data["docs"] = docs
        await loader.documents_to_json(
            documents=data["docs"], filename=output_path / f"{k}_raw.json"
        )
        data["cleaned_docs"] = await cleaner.clean_documents(documents=docs)
        await loader.documents_to_json(
            documents=data["cleaned_docs"], filename=output_path / f"{k}_clean.json"
        )
```
